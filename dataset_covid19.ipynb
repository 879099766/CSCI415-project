```python
# check the missing val in the dataset

# get the number of missing data points per column
missing_values_count = data.isnull().sum()

# how many total missing values do we have?
total_cells = np.product(data.shape)
total_missing = missing_values_count.sum()

# percent of data that is missing
percent_missing = (total_missing/total_cells) * 10
print(percent_missing)

# in this case, the missing val was not available
# by looking at the missing val attributes
```

```python
# handling missing val
# by filling 'NaN' with 0

# NOTE: make sure you run this on GPU or cloud!!!
data.fillna(0, inplace=True)
```

```python
country_grp = data.groupby(["location"]).sum().reset_index()
country_grp.sort_values(["total_cases"], ascending=False, inplace=True)
country_grp["total_cases"] = country_grp["total_cases"].astype(int)

global country_list
country_list = country_grp.iloc[:]["location"].unique().tolist()

# remove two unused val
country_list.remove("International")
country_list.remove("World")
```

```python
def form_raceChart_df():
    date_list = data["date"].unique()
    date_list.sort()

    # fill missing val with 0
    filler = 0

    global race_total_case
    race_total_case = pd.DataFrame()

    for country in country_list:
        filt = (data["location"] == str(country))
        x = data.loc[filt]
        y = x["total_cases"]
        z = y.tolist()
        # fill the actual data from the descending order
        # fill 0 in ascending order
        race_total_case.loc[:, country] = [filler]*(len(race_total_case.index) - len(z)) + z

    # set the index for the tbl for visualization
    race_total_case.index = date_list
```

```python
import warnings
```

```python
def form_raceChart():
    # hide all warnings from the missing font
    warnings.filterwarnings('ignore')

    bcr.bar_chart_race(
        df = race_total_case,
        filename = "test.mp4",
        steps_per_period = 10,
        dpi = 144,
        title = 'COVID-19 Top 10 Countries',
        n_bars = 10,
        filter_column_colors = True,
        figsize = (5, 3),
        )

    # restore the warning
    warnings.filterwarnings("default")

    print("Output done!")
```
