```python
d_parser = lambda x: datetime.strptime(x, '%Y-%m-%d')

data = pd.read_csv(
    "https://covid.ourworldindata.org/data/owid-covid-data.csv", 
    parse_dates=['date'],
    date_parser=d_parser,
                  )
# data = pd.read_csv("owid-covid-data.csv")
```

```python
# check the missing val in the dataset

# get the number of missing data points per column
missing_values_count = data.isnull().sum()

# how many total missing values do we have?
total_cells = np.product(data.shape)
total_missing = missing_values_count.sum()

# percent of data that is missing
percent_missing = (total_missing/total_cells) * 100
print(percent_missing)

# in this case, the missing val was not available
# by looking at the missing val attributes
```

```python
# handling missing val
# by filling 'NaN' with 0

# NOTE: make sure you run this on GPU or cloud!!!
data.fillna(0, inplace=True)
```

```python
# ============== Top 10 most infected countries ==============


country_grp = data.groupby(["location"]).sum().reset_index()
country_grp.sort_values(["total_cases"], ascending=False, inplace=True)
country_grp["total_cases"] = country_grp["total_cases"].astype(int)

global country_list
country_list = country_grp.iloc[:]["location"].unique().tolist()

# remove two unused val
country_list.remove("International")
country_list.remove("World")
```

```python
def form_raceChart_df():
    date_list = data["date"].unique()
    date_list.sort()

    # fill missing val with 0
    filler = 0

    global race_total_case
    race_total_case = pd.DataFrame()

    for country in country_list:
        filt = (data["location"] == str(country))
        x = data.loc[filt]
        y = x["total_cases"]
        z = y.tolist()
        # fill the actual data from the descending order
        # fill 0 in ascending order
        race_total_case.loc[:, country] = [filler]*(len(race_total_case.index) - len(z)) + z

    # set the index for the tbl for visualization
    race_total_case.index = date_list
```

```python
import warnings
```

```python
def form_raceChart():
    # hide all warnings from the missing font
    warnings.filterwarnings('ignore')

    bcr.bar_chart_race(
        df = race_total_case,
        filename = "Top_10_covid.mp4",
        steps_per_period = 10,
        dpi = 144,
        title = 'Top 10 COVID-19 in fected Countries',
        n_bars = 10,
        filter_column_colors = True,
        figsize = (5, 3),
        )

    # restore the warning
    warnings.filterwarnings("default")

    print("Output done!")
```

```python
form_raceChart_df()
form_raceChart()
```

```python
# ============== Monthly case by country ==============

global country
target_country = "United States"
filt = (data["location"] == target_country)
global case_Country
case_Country = data.loc[filt]
# case_Country
```

```python
def get_lastDay_month():
    temp_date = case_Country.groupby('location').resample('M', on='date').sum().reset_index().sort_values(by='date')
    global month_dateList
    month_dateList = temp_date["date"]
    month_dateList.tolist()
    month_dateList = month_dateList.apply(lambda x : x.strftime('%Y-%m-%d'))
    # remove current end of month
    month_dateList = month_dateList[:-1]
    
def get_monthly_case():
    global monthly_case_country
    temp = []
    monthly_case_country = []

    monthly_case_country.append(0)
    
    for val in month_dateList:
        temp.append(case_Country.loc[case_Country['date'] == val]['total_cases'].astype(int).values)
    
    for i in temp:
        for j in i:
            monthly_case_country.append(j)

def form_xLabel():
    temp_df = case_Country.groupby('location').resample('M', on='date').sum().reset_index().sort_values(by='date')
    
    global x_label
    x_label = temp_df["date"]
    x_label.tolist()
    x_label = x_label.apply(lambda x: x.strftime('%m/%d'))
    
def cal_monthly_activeCase():
    # calculate active case
    # active case = total_case - (total_death + total_recovery)
    # yet, this dataset doesn't provide recovery case attribute
    global active_case
    x = []
    temper = []
    active_case = []
    counter = 0
    
    active_case.append(0)
    
    for val in month_dateList:
        x.append(case_Country.loc[case_Country['date'] == val]['total_deaths'].astype(int).values)
    
    for i in x:
        for j in i:
            temper.append(j)
    
    for item in temper:
        active_case.append(monthly_case_country[counter] - temper[counter])
        counter = counter + 1

        
def cal_monthly_death():
    global death_case
    temp = []
    death_case = []
    
    death_case.append(0)
    
    for val in month_dateList:
        temp.append(case_Country.loc[case_Country['date'] == val]['total_deaths'].astype(int).values)
    
    for i in temp:
        for j in i:
            death_case.append(j)

def cal_monthly_recovery():
    # calculate monthly recovery cases
    # formular: total - active - death = recovery
    global case_recovery
    case_recovery = []
#     case_recovery.append(0)
    
    for x in range(0, len(monthly_case_country)):
        case_recovery.append(
            (monthly_case_country[x]) - (active_case[x] + death_case[x])
        )
            
def form_lineGraph():
    plt.plot(x_label, monthly_case_country, linestyle="-", color='#444444', label='Total Cases')
    plt.plot(x_label, active_case, linestyle="--", color='blue', label="Active Cases")
    plt.plot(x_label, death_case, linestyle=":", color='red', label="Death Cases")
    plt.plot(x_label, case_recovery, linestyle="dashed", color='green', label="Recovery Cases")

    # second_axes = plt.twinx() # create the second axes, sharing x-axis
    # second_axis.set_yticks(active_case) # list of your y values
    # pyplot.show() # update the figure

#     plt.fill_between(x_label, active_case, case_recovery, 
#              where=(case_recovery > active_case), 
#              interpolate=True, color='green',alpha=0.25, label='High Recovery')
        
#     plt.fill_between(x_label, case_recovery, active_case,
#                  where=(active_case <= case_recovery), 
#                  interpolate=True, color='red', alpha=0.25, label='Low Recovery')
    plt.legend()
    plt.title("Types of cases in US")
    plt.xlabel('Month')
    plt.ylabel('Cases')
#     plt.tight_layout()
    
    plt.tight_layout()
    # 1e8 is standard scientific notion, and here it indicates an overall scale factor for the y-axis. 
    # That is, if there's a 2 on the y-axis and a 1e8 at the top, 
    # the value at 2 actually indicates 2*1e8 = 2e8 = 2 * 10^8 = 200,000,000.

    # Ref: https://stackoverflow.com/questions/25715333/in-python-plots-what-does-1e8-mean-in-reference-to-the-y-axis
```

```python
get_lastDay_month()
get_monthly_case()
cal_monthly_activeCase()
cal_monthly_death()
cal_monthly_recovery()
form_xLabel()
form_lineGraph()
```

```python
# ============== gdp_per_capita is affected by COVID-19 ==============


# refer to second visualization val
# "case_Country" contains the country we wanna explor

def get_monthly_gdp_per_capita():
    global gdp_per_capita
    temp = []
    gdp_per_capita = []

    gdp_per_capita.append(0)
    
    for val in month_dateList:
        temp.append(case_Country.loc[case_Country['date'] == val]['gdp_per_capita'].astype(int).values)
    
    for i in temp:
        for j in i:
            gdp_per_capita.append(j)

def form_lineGraph_gdp_per_capita():
    plt.plot(x_label, monthly_case_country, linestyle="-", color='#444444', label='Total Cases')
    plt.plot(x_label, gdp_per_capita, linestyle="--", color='blue', label="GPD Per Capital")
    plt.legend()
    plt.title("GPD per Capital V.S. COVID-19 Cases (monthly)")
    plt.xlabel('Month')
    plt.ylabel('Value')
    plt.tight_layout()
```

```python
get_monthly_case()
get_lastDay_month()
form_xLabel()
get_monthly_gdp_per_capita()
form_lineGraph_gdp_per_capita()
```

```python
conts = data["continent"].unique()

def form_y_label():
    global y_label
    y_label = []
    
    for cont in conts:
        if cont == 0:
            cont == "Unidentified"
        y_label.append(cont)

        
def get_continent_grp():
    
    today = datetime.today()
    today = today.strftime("%Y-%m-%d")
    
    filt = (data["date"] == today)
    today_conts = data[filt]
    # today_conts

    global cont_grp
    cont_grp = today_conts.groupby("continent").sum().reset_index()
    cont_grp = cont_grp[1:]


def form_barGraph_cont():
    plt.barh(cont_grp["continent"], cont_grp["total_cases"])
    plt.title("COVID-19 Total Cases Per Continent")
    plt.ylabel("Continents")
    plt.xlabel("Number of Cases")
    plt.tight_layout()
```

```python
get_continent_grp()
form_barGraph_cont()
```
